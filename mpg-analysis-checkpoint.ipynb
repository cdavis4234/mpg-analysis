```python
import seaborn as sns
sns.load_dataset("mpg")
```

```python
mpg = sns.load_dataset("mpg")
mpg
```

```python
# According to the visual below, the vehicles with the highest horsepower have the lowest mpg on average, 
# while the vehicles with the lowest horsepower have the highest mpg
sns.relplot(data=mpg,x="horsepower",y="mpg")
```

```python
# Japan cars have the highest mpg on average, while the USA has the lowest
mpg.groupby("origin").mpg.mean().sort_values(ascending=False)
```

```python
# Over time, the fuel economy has increased with each model year. In 1970, the average mpg was around 18. In 1982 it is around 31 mpg. 
# The fuel economy has gone up and down with each model year, but the trend has been upward.
sns.lineplot(data=mpg,x="model_year",y="mpg", errorbar=None, color='purple')
```

```python
# In the visual below, I compared the car's weight and the mpg.
# I noticed that the more the car weighs, the lower the mpg, and the lighter it is, the higher the mpg is.
sns.relplot(data=mpg,x='weight',y='mpg', color = 'red')
```

```python
# In the visual below, I compared the mpg with the car's country of origin.
# The USA clearly has the lowest mpg for cars manufactured, while Japan has the highest
sns.catplot(data=mpg,x="origin",y="mpg",kind="bar", estimator='mean')
```
